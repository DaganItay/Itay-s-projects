```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
df = pd.read_excel("C:\\Users\\dagani\\OneDrive - HP Inc\\Itay General\\Machine learning\\DFfinal.xlsm")
X = df.drop(['y'], axis=1)

Y=df.groupby('SN').mean('y')
y=(Y['y']).astype(int)
y.value_counts()/len(y)
```

```python
y
```

```python
from tsfresh import extract_features
from tsfresh.feature_extraction import MinimalFCParameters
from tsfresh.feature_extraction import ComprehensiveFCParameters

settings = MinimalFCParameters()
#settings = ComprehensiveFCParameters()

extracted_features = extract_features(X,column_id='SN', column_sort='time',default_fc_parameters=settings)
```

```python
extracted_features
```

```python
index=extracted_features.index
X = preprocessing.StandardScaler().fit_transform(extracted_features)
extracted_features = pd.DataFrame(X, columns=extracted_features.columns)
extracted_features.index=index
```

```python
extracted_features
```

```python
from sklearn import model_selection

X_train, X_test, y_train, y_test = model_selection.train_test_split(extracted_features, y, test_size=0.30, stratify=y, random_state=42)

len(X_train), len(X_test)
```

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

model = KNeighborsClassifier()
model.fit(X_train, y_train)
y_preds = model.predict(X_test)

print(metrics.confusion_matrix(y_test, y_preds))
print(metrics.classification_report(y_test, y_preds))
```

```python
from sklearn import model_selection



from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = model_selection.train_test_split(extracted_features, y, test_size=0.20, stratify=y, random_state=42)
model = RandomForestClassifier(n_estimators=50, random_state=42)
model.fit(X_train, y_train)

y_preds = model.predict(X_test)

print(metrics.confusion_matrix(y_test, y_preds))
print(metrics.classification_report(y_test, y_preds))
```
